```python
import pandas as pd
import numpy as ny
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
dataset=pd.read_csv("INC.csv")
```

```python
Quality=dataset['sex']
dataset['sex']=dataset['sex'].map({'female':0,'male':1}).astype(int)
```

```python
CONVERT=dataset["smoker"]
dataset['smoker']=dataset['smoker'].map({'yes':1,'no':0}).astype(int)
```

```python
x=dataset.drop('charges',axis=1)
```

```python
y=dataset.charges
y
```

```python
dataset['bmi'] = dataset['bmi'].astype(int)
```

```python
origin=x['region']
x['region']=x['region'].map({'southwest':0,'southeast':1,'northwest':2,'northeast':3}).astype(int)
```

```python
x['bmi'] = x['bmi'].round().astype(int)
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.24,random_state=0)
```

```python
from sklearn.svm import SVR
model=SVR()
model.fit(x_train,y_train)
```

```python
y_pred=model.predict(x_test)
from sklearn.metrics import r2_score,mean_squared_error
mse = mean_squared_error(y_test,y_pred)
print(mse)
```

```python
rmse=ny.sqrt(mse)
print("Root Mean Error:",rmse)
r2_score=r2_score(y_test,y_pred)
print("R2score:",r2_score)
```
